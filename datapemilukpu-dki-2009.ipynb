```python
# Import pandas untuk membaca file Excel
import pandas as pd
# Import library yang diperlukan
import matplotlib.pyplot as plt
import seaborn as sns
# Import library yang diperlukan
import matplotlib.pyplot as plt
import seaborn as sns

# Membaca data Excel
file_path = '/content/drive/My Drive/Colab Notebooks/DataMining/datapemilukpu-dki-2009.xls'
data = pd.read_excel(file_path)

```

```python
# Mengelompokkan data berdasarkan status terpilih atau tidak
terpilih = data[data['Terpilih atau Tidak'] == 'YA']
tidak_terpilih = data[data['Terpilih atau Tidak'] == 'TIDAK']
```

```python
# Membuat boxplot untuk membandingkan suara sah calon
plt.figure(figsize=(10,6))
sns.boxplot(x='Terpilih atau Tidak', y='Suara Sah Caleg', data=data)
plt.title('Perbandingan Jumlah Suara Sah Antara Caleg Terpilih dan Tidak Terpilih')
plt.show()
```

```python
# Statistik deskriptif untuk calon terpilih
print("Statistik deskriptif calon terpilih:")
print(terpilih['Suara Sah Caleg'].describe())

# Statistik deskriptif untuk calon tidak terpilih
print("\nStatistik deskriptif calon tidak terpilih:")
print(tidak_terpilih['Suara Sah Caleg'].describe())
```
